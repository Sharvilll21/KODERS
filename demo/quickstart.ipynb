```python
from src.benchmarking import compare_functions
import numpy as np

data = np.random.rand(1000000)

def loop_sum(x):
    total = 0
    for val in x:
        total += val
    return total

def vectorized_sum(x):
    return np.sum(x)

compare_functions(vectorized_sum, loop_sum, data)
```
